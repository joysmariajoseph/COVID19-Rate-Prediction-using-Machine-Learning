```python
import pandas as pd
data=pd.read_csv("owid-covid-data.csv")
```

```python
type(data)
```

```python
data.info()
```

```python
data=data[data["location"]=="India"]
data.reset_index(drop=True, inplace=True)
```

```python
data["new_tests"].fillna(data["new_tests"].mean(),inplace=True)
data["total_tests"].fillna(data["total_tests"].mean(),inplace=True)
data["total_tests_per_thousand"].fillna(data["total_tests_per_thousand"].mean(),inplace=True)
data["new_tests_per_thousand"].fillna(data["new_tests_per_thousand"].mean(),inplace=True)
data["new_tests_smoothed"].fillna(data["new_tests_smoothed"].mean(),inplace=True)
data["new_tests_smoothed_per_thousand"].fillna(data["new_tests_smoothed_per_thousand"].mean(),inplace=True)
data["tests_per_case"].fillna(data["tests_per_case"].mean(),inplace=True)
data["positive_rate"].fillna(data["positive_rate"].mean(),inplace=True)
data["stringency_index"].fillna(data["stringency_index"].mean(),inplace=True)
```

DATE CONVERSION

```python
data['date'] = pd.to_datetime(data['date'])
import datetime as dt
data['date']=data['date'].map(dt.datetime.toordinal)
data['date']
```

```python
data.info()
```

```python
X=data.drop(["total_cases","tests_units","iso_code","continent","location"],axis=1)
Y=data['total_cases']
# print(Y)
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=52)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,Y_train)
```

```python
X_test.shape
```

```python
LR.predict(X_test)
```

```python
LR.score(X_test,Y_test)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
reg_cv=LinearRegression()
cv_scores=cross_val_score(reg_cv,X,Y,cv=6)
np.mean(cv_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR=RandomForestRegressor()
RFR.fit(X_train,Y_train)
RFR.predict(X_test)
```

```python
RFR.score(X_test,Y_test)
```
